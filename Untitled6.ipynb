```python
import pandas as pd

oceania = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
```

```python
years = oceania.columns.str.strip('gdpPercap_')
years
```

```python
oceania.columns = years.astype(int)
```

```python
type(oceania.columns[0])
```

```python
oceania.loc['Australia'].plot()
```

```python
oceania.loc['Australia']
```

```python
oceania.T.plot()
plt.ylabel('GDP per capita')
```

```python
oceania.plot()
```

```python
plt.style.use('ggplot')
oceania.T.plot(kind='bar')
plt.ylabel('GDP per capita');  # <<--- trick: add a semicolon to make cleaner output
```

```python
gdp_australia = oceania.loc['Australia']
gdp_australia
```

```python
years = oceania.columns
years
```

```python
plt.plot(years, gdp_australia, 'g--');
```

```python
gdp_nz = oceania.loc['New Zealand']
plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label='New Zealand')
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita');
```

```python
plt.scatter(gdp_australia, gdp_nz)
```
